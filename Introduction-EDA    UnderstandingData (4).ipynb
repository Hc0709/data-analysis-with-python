## Filling MissingValues - Imputations
i) Numerical data-use Mean

```python
df3=df.copy()
```

```python
df3
```

```python
df3.isnull().sum()
```

```python
#before filling null values
df.iloc[:,13:14]
```

```python
mean_val = df3['reviews_per_month'].mean()
df3['reviews_per_month'].fillna(mean_val,inplace=True)
```

```python
df3.head()
```

```python
#After filling null values
df3.iloc[:,13:14]
```

```python
df3.isnull().sum()
```

ii) Filling CategoricalData use mode

```python
df3.nunique()
```

```python
df3["last_review"]
```

```python
df3.isnull().sum()
```

```python
df3["last_review"].fillna(df3["last_review"].value_counts().index[0],inplace=True)
```

```python
df3["last_review"]
```

```python
df3.isnull().sum()
```

```python
df4 =df.copy()
```

```python
df4.isnull().sum()
```

## Filling our own value

```python
df4["last_review"].fillna('not_reviewed',inplace =True)
```

```python
df4["last_review"]
```

## Advanced imputations
interpolate(gives value based on surrounding values

```python
df5.isnull().sum()
```

```python
df5["reviews_per_month"]
```

```python
df5["reviews_per_month"].interpolate(inplace = True)
```

## Dealing with Duplicate Data

```python

import pandas as pd
df6 =pd.read_csv('googleplaystore.csv')
```

```python
df6.duplicated().sum()
```

```python
df6.shape
```

```python
df6[df6.duplicated()]
```

```python
df6[df6.duplicated(keep="first")]  #keeps first duplicated rows and returns last duplicated rows
```

```python
df6[df6.duplicated(keep="last")] #keeps last duplicated rows and returns first duplicated rows
```

```python
df6[df6.duplicated(keep = "False")] #displays all duplicated values
```

```python
df6.drop_duplicates()
```

```python
df6.duplicated().sum()
```

```python
df6.duplicated("Rating").sum()
```

```python
df6.drop_duplicates("Rating",keep="last")
```

## Null values- Categorical

```python
#country column
df.iloc[:,:1]
```

```python
#filling null values instead of removing it(country column)
imputer = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
imputer.fit(df.iloc[:,:1].values)
df.iloc[:,:1]=imputer.transform(df.iloc[:,:1].values)
```

```python
df
```

```python
#purchased column
df.iloc[:,3:]
```

```python
imputer = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
imputer.fit(df.iloc[:,3:].values)
df.iloc[:,3:]=imputer.transform(df.iloc[:,3:].values)
```

```python
df
```

```python
#replacing entire data null values at once
imputer = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
imputer.fit(df.iloc[:,:].values)
df.iloc[:,:]=imputer.transform(df.iloc[:,:].values)
```

```python
df
```

```python
#replacing entire data null values at once(i loc for particular columns replacement only)
imputer = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
imputer.fit(df)
df=imputer.transform(df)
```
